```python
import numpy,pandas
from PIL import Image, ImageFilter
from scipy.ndimage import convolve
import os,cv2

shape,color=(512,512),cv2.COLOR_BGR2HSV #COLOR_BGR2HSV

def LoG_filter(image, sigma, size=None):
    if size is None:
        size = int(6 * sigma + 1) if sigma >= 1 else 7
    if size % 2 == 0:
        size += 1
    x, y = numpy.meshgrid(numpy.arange(-size//2+1, size//2+1), numpy.arange(-size//2+1, size//2+1))
    kernel = -(1/(numpy.pi * sigma**4)) * (1 - ((x**2 + y**2) / (2 * sigma**2))) * numpy.exp(-(x**2 + y**2) / (2 * sigma**2))
    kernel = kernel / numpy.sum(numpy.abs(kernel))
    result = convolve(image, kernel)
    return result

train_info=pandas.read_csv("/kaggle/input/ai-vs-human-generated-dataset/train.csv")

if os.path.exists(f"/kaggle/working/train/")==False:
    os.mkdir(f"/kaggle/working/train/")
    os.mkdir(f"/kaggle/working/train/real/")
    os.mkdir(f"/kaggle/working/train/editada/")

for i in zip(train_info['file_name'],train_info['label'],range(len(train_info['file_name']))):
    #image = Image.open(f"/kaggle/input/ai-vs-human-generated-dataset/{i[0]}")
    image = cv2.cvtColor(cv2.resize(cv2.imread(f"/kaggle/input/ai-vs-human-generated-dataset/{i[0]}"), shape), color)

    if i[1]==0:
        #image.filter(filter=ImageFilter.EDGE_ENHANCE_MORE).save(f"/kaggle/working/train/real/{i[0].split('/')[1]}")
        cv2.imwrite(f"/kaggle/working/train/real/{i[0].split('/')[1]}", image)# cv2.cvtColor(image, color)) LoG_filter(image, 1.0)
    elif i[1]==1:
        #image.filter(filter=ImageFilter.EDGE_ENHANCE_MORE).save(f"/kaggle/working/train/editada/{i[0].split('/')[1]}")
        cv2.imwrite(f"/kaggle/working/train/editada/{i[0].split('/')[1]}", image)# cv2.cvtColor(image, color)) LoG_filter(image, 1.0)
```
